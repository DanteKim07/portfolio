#1. Sentence_transformers 다운로드


```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ['반갑습니다', '한국어를 위한 임베딩 버트모델입니다.']
embeddings = model.encode(sentences)

print(embeddings)
```

#2. 약 챗봇에 필요한 데이터파일 업로드 및 임베딩

```python
drug_chat = pd.read_csv('이약저약_챗봇2.csv')#, encoding='utf-8')

drug_chat.head()
```

```python
drug_chat.loc[3,'채팅 내용']
```

```python
drug_chat['embedding'] = pd.Series([[]] * len(drug_chat)) #더미

drug_chat['embedding'] = drug_chat['채팅 내용'].map(lambda x: list(model.encode(x)))

drug_chat.head()
```

```python
drug_chat.to_csv('drug_chatbot.csv', index=False)
```

#3. 채팅 적용

```python
text = '우울증약이랑 진통제 같이 먹어도 되나요?'

embedding_text = model.encode(text)
```

```python
drug_chat['유사도'] = drug_chat['embedding'].map(lambda x: cosine_similarity([embedding_text], [x]).squeeze())

drug_chat.head()
```

```python
answer = drug_chat.loc[drug_chat['유사도'].idxmax()]

print('약 이름: ', answer['약 구분'])
print('비슷한 질문: ',answer['채팅 내용'])
print('챗봇 응답: ',answer['답변'])
print('유사도: ',answer['유사도'])
```
