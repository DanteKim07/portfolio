```python
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', '', str(text))
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          '', cleaned_text)
    cleaned_text = re.sub(' +', ' ', cleaned_text)
    
    return cleaned_text
```

```python
title2 = []
warn = []
sunb2 =[]
sha2 = []
hyo2 =[]
yong2 =[]
gy2 = []
bok2 =[]

for i in new_d_str2:
    driver.get(f'https://www.health.kr/drug_info/_kpanet2/basedrug/show_detail.asp?table=kdrug_basedrug.dbo.basedrug&idx={i}')
    time.sleep(2)
    title = driver.find_elements_by_css_selector('body > table > tbody > tr:nth-child(1) > td > table:nth-child(1) > tbody > tr:nth-child(1) > td:nth-child(2) > b > font')
    
    #이름
    for j in title:
        title2.append(j.text)
        
    
    # 성분명
    sunb = driver.find_elements_by_id('t1')
    for k in sunb:
        sunb2.append(k.text)
    
    # 성상
    sha = driver.find_elements_by_id('t2')
    for k2 in sha:
        sha2.append(k2.text)
    
    #효능, 효과
    hyo = driver.find_elements_by_id('t3_2')
    for k3 in hyo:
        hyo2.append(k3.text)
    
    #용법, 용량
    yong = driver.find_elements_by_id('t4_2')
    for k4 in yong:
        yong2.append(k4.text)
    
    #경고
    gy = driver.find_elements_by_id('t5_2')
    for k5 in gy:
        gy2.append(k5.text)
    
    # 복약지도
    bok = driver.find_elements_by_id('t7')
    for k6 in bok:
        bok2.append(k6.text)
        
    driver.back()
    time.sleep(2)
    driver.switch_to.frame(driver.find_element(By.XPATH,'//*[@id="contents"]/div[1]/div[2]/div[2]/iframe'))
    
```
